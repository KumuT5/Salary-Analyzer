```python
#removing the metadata
ori1=pd.read_csv("2019responses.csv",header=3,index_col=0)
ori2=pd.read_csv("2021responses.csv",header=3,index_col=0)

#printing the data file
print("data without metadata")
display(ori1.head())

print("data without metadata")
display(ori2.head())
```

CHECK & CLEAN DATASET

```python
#adding column year
ori1["Survey_Year"]=2019
ori2["Survey_Year"]=2021

#row n column count
print("count:\t",ori1.shape)
print("count:\t",ori2.shape)

# merging the two CSVs for a better view
dataset=pd.concat([ori1,ori2],ignore_index=True)

# Drop missing or zero salary rows

print("\n",dataset.columns.tolist())

dataset = dataset.drop(columns=["Survey Year"])

print("\nUpdated List Of Column\n")
print(dataset.columns.tolist())

# Convert SalaryUSD column to numeric
dataset["SalaryUSD"] = pd.to_numeric(dataset["SalaryUSD"], errors="coerce")

if "SalaryUSD" in dataset.columns:
    dataset = dataset[dataset["SalaryUSD"].notna()]
    dataset = dataset[dataset["SalaryUSD"] > 0]


#final row n column count & average salary
print("\nRow & Column Count")
print(dataset.shape)
print("\nAVERAGE SALARY")
dataset.groupby("Survey_Year")["SalaryUSD"].mean()
```

```python
#drop unrequired data
del_cols=[
    'Timestamp', 'PostalCode', 'PrimaryDatabase', 'YearsWithThisDatabase',
    'OtherDatabases', 'DatabaseServers', 'NewestVersionInProduction',
    'OldestVersionInProduction', 'Certifications', 'CareerPlansThisYear',
    'LookingForAnotherJob', 'Gender', 'OtherJobDuties', 'KindsOfTasksPerformed',
    'Counter', 'ManageStaff', 'TeamSize','PopulationOfLargestCityWithin20Miles','HowManyCompanies'
]
dataset=dataset.drop(columns=[c for c in del_cols if c in dataset.columns])

print(dataset.columns.tolist())

#rename the columns
rename={
    "SalaryUSD": "Salary",
    'EmploymentStatus': 'Employment',
    'JobTitle': 'JobTitle',
    'YearsWithThisTypeOfJob': 'Experience',
    'OtherPeopleOnYourTeam': 'TeamSize',
    'CompanyEmployeesOverall': 'CompanySize',
    'Education': 'Education',
    'EducationIsComputerRelated': 'CS_Education',
    'HoursWorkedPerWeek': 'WorkHours',
    'TelecommuteDaysPerWeek': 'RemoteDays',
    'EmploymentSector': 'Sector',
    'Survey_Year': 'Year'
}
dataset=dataset.rename(columns=rename)
print("\nUpated list of columns\n")
print(dataset.columns.tolist())      

```

```python
print(dataset.duplicated().any())
```

```python
dataset=dataset.drop_duplicates()

print(dataset.duplicated().any())
```

```python
print(dataset.dtypes)
```

```python
# columns that should be numeric
num = ['TeamSize', 'WorkHours', 'RemoteDays']

for c in num:
    dataset[c] = pd.to_numeric(dataset[c], errors='coerce')

to_cat = ['Country','Employment','JobTitle','CompanySize','Education','CS_Education','Sector']

for c in to_cat:
    dataset[c] = dataset[c].astype('category')

print(dataset.dtypes)
    
```

```python
#checking for null values
print(dataset.isnull().sum())
```

```python
col1=['Salary', 'Experience','TeamSize','WorkHours','RemoteDays']
for c in col1:
    dataset[c]=dataset[c].fillna(dataset[c].median())

col2=['Country', 'Employment', 'JobTitle', 'CompanySize','Education','Sector']
for a in col2:
    dataset[a]=dataset[a].fillna(dataset[a].mode()[0])

print(dataset['CS_Education'].unique())    

col3=dataset['CS_Education'] = dataset['CS_Education'].fillna('Not Asked')

# Remove extreme outliers (IQR method)
if "Salary" in dataset.columns:
    Q1 = dataset["Salary"].quantile(0.25)
    Q3 = dataset["Salary"].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    dataset = dataset[(dataset["Salary"] >= lower) & (dataset["Salary"] <= upper)]

```

```python
#checking for null values
print(dataset.isnull().sum())
```

```python
dataset = dataset[(dataset != 0).all(axis=1)]
dataset = dataset.dropna(how='all')
```

SCATTER PLOT SAL VS EXP.

```python
# Fix year-like Experience values and re-plot salary vs experience
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------- 1) find relevant columns ----------
def find_col(names):
    for c in dataset.columns:
        if any(n.lower() == c.lower() for n in names):
            return c
    return None

exp_col = find_col(["experience","yearswiththistypeofjob","yearsofexperience"])
year_col = find_col(["year","survey_year","surveyyear"])
salary_col = find_col(["salary","salaryusd"])

print("Detected columns -> Experience:", exp_col, "| Year:", year_col, "| Salary:", salary_col)

# ---------- 2) quick diagnostics BEFORE cleaning ----------
print("\nBefore cleaning - Experience summary (top problematic values):")
if exp_col:
    print(dataset[exp_col].describe())
    print("Top 10 largest experience values:\n", dataset[exp_col].sort_values(ascending=False).head(10))

print("\nSalary summary (quick):")
if salary_col:
    print(dataset[salary_col].describe())
    print("Top 10 salaries:\n", dataset[salary_col].sort_values(ascending=False).head(10))

# ---------- 3) clean Experience: convert year-like entries into duration ----------
ds = dataset.copy()

# coerce to numeric (safe)
if exp_col:
    ds[exp_col] = pd.to_numeric(ds[exp_col], errors="coerce")

# define fallback survey year if dataset doesn't have Year column
fallback_survey_year = None
if year_col and year_col in ds.columns:
    # pick most common Year value or max if missing
    try:
        fallback_survey_year = int(ds[year_col].mode(dropna=True)[0])
    except Exception:
        fallback_survey_year = int(ds[year_col].max(skipna=True))
else:
    # you used 2019 & 2021 earlier; use the latest available as fallback
    fallback_survey_year = 2021

print("\nUsing fallback_survey_year =", fallback_survey_year)

# convert values that look like years ( > 100 and <= fallback_survey_year ) to experience
if exp_col:
    mask_year_like = ds[exp_col].notna() & (ds[exp_col] > 100) & (ds[exp_col] <= fallback_survey_year)
    print("Year-like Experience count:", mask_year_like.sum())
    # convert year -> experience: survey_year - year_value
    ds.loc[mask_year_like, exp_col] = fallback_survey_year - ds.loc[mask_year_like, exp_col]

# After conversion coerce again to numeric
if exp_col:
    ds[exp_col] = pd.to_numeric(ds[exp_col], errors='coerce')

# Clamp unrealistic values: negative or > 60 -> set to NaN
if exp_col:
    bad_mask = (ds[exp_col] < 0) | (ds[exp_col] > 60)
    print("Values flagged as unrealistic (to be set NaN):", bad_mask.sum())
    ds.loc[bad_mask, exp_col] = np.nan

# Fill missing Experience with median (safe beginner choice)
if exp_col:
    med_exp = ds[exp_col].median(skipna=True)
    print("Filling missing Experience with median:", med_exp)
    ds[exp_col] = ds[exp_col].fillna(med_exp)

# ---------- 4) ensure Salary numeric & no accidental scaling ----------
if salary_col:
    # strip non-numeric chars then coerce
    ds[salary_col] = pd.to_numeric(ds[salary_col].astype(str).str.replace(r'[^0-9.]', '', regex=True), errors='coerce')
    # quick check: if median salary is tiny (<1000) while max > 1000 then we have mixed units;
    # We will not auto-rescale, but we print hint for you.
    median_salary = ds[salary_col].median(skipna=True)
    max_salary = ds[salary_col].max(skipna=True)
    print("\nPost-clean Salary stats -> median:", median_salary, " max:", max_salary)
    if median_salary is not None and median_salary < 1000:
        print("NOTE: median salary < 1000. If salaries are in lakhs/1000s you may want to rescale (e.g., *100 or *1000).")

# ---------- 5) Diagnostics AFTER cleaning ----------
print("\nAfter cleaning - Experience summary:")
if exp_col:
    print(ds[exp_col].describe())
    print("Top experiences now:\n", ds[exp_col].sort_values(ascending=False).head(10))

# ---------- 6) Replace dataset or keep as ds ----------
# if you'd like to keep original and update dataset variable:
dataset = ds  # update the main variable for subsequent plotting

# ---------- 7) re-plot clean Salary vs Experience ----------
plt.figure(figsize=(10,6))
sns.set(style="whitegrid")
sns.regplot(
    data=dataset,
    x=exp_col,
    y=salary_col,
    scatter_kws={"alpha":0.5, "s":40},
    line_kws={"color":"red", "linewidth":2},
    ci=95
)
plt.title("Salary vs Experience (cleaned)", fontsize=14, weight="bold")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.tight_layout()
plt.show()
```

COLUMN CHART SAL VS JOB TITLES

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Compute median salary for each Job Title
job_salary = (
    dataset.groupby('JobTitle')['Salary']
    .median()
    .sort_values(ascending=False)
    .head(10)        # Top 10 highest-paying job titles
)

# Create a DataFrame for plotting
job_df = pd.DataFrame({
    'JobTitle': job_salary.index.astype(str),
    'MedianSalary': job_salary.values
})

# Plot
plt.figure(figsize=(12,6))
sns.set(style="whitegrid")

sns.barplot(
    data=job_df,
    x='JobTitle',
    y='MedianSalary',
    palette='viridis'
)

plt.title("Top 10 Job Titles by Median Salary", fontsize=16, weight='bold')
plt.xlabel("Job Title", fontsize=12)
plt.ylabel("Median Salary (USD)", fontsize=12)

# Rotate job titles for readability
plt.xticks(rotation=45, ha='right')

# Add values on bars
for i, row in job_df.iterrows():
    plt.text(i, row['MedianSalary'], f"{row['MedianSalary']:,.0f}",
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = ["Experience", "WorkHours", "RemoteDays", "TeamSize"]

plt.figure(figsize=(14,10))
sns.set_style("white")

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=dataset, x=col, bins=30, color="steelblue", kde=True)
    sns.despine(trim=True)
    plt.title(f"{col} Distribution", fontsize=14, weight='bold')

plt.tight_layout()
plt.show()
```

PIE CHART EDUCATION LEVEL

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
edu_counts = dataset['Education'].value_counts()

# Pretty color palette (soft blues + purples)
colors = sns.color_palette("coolwarm", n_colors=len(edu_counts))

# Create the figure
plt.figure(figsize=(8,8))

# Pie chart with improved styling
wedges, texts, autotexts = plt.pie(
    edu_counts.values,
    labels=edu_counts.index,
    autopct='%1.1f%%',
    pctdistance=0.8,
    labeldistance=1.1,
    startangle=140,
    colors=colors,
    textprops={'fontsize': 12}
)

# Style percentage text
for autotext in autotexts:
    autotext.set_color("white")
    autotext.set_fontsize(12)
    autotext.set_weight("bold")

# Remove black edges for a soft look
for w in wedges:
    w.set_linewidth(0)

# Add title
plt.title("Education Level Distribution", fontsize=18, weight='bold', pad=20)

# Perfect layout
plt.tight_layout()
plt.show()
```
